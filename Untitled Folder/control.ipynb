```python
from __future__ import print_function
from __future__ import division
import numpy as np
from matplotlib import pyplot as plt
from math import cos, sin, pi, atan2, sqrt
```

```python
import pylab # this allows you to control figure size 
pylab.rcParams['figure.figsize'] = (10.0, 8.0) # this controls figure size in the notebook
```

```python
def drive(phi, v):
    # v - max speed
    v /=sqrt(2)
    speed = np.array([v,v])
    M = np.array([
        [cos(phi), -sin(phi)],
        [sin(phi), cos(phi)]
    ])
    return M.dot(speed)
```

```python
dt = 8
for i in range(dt):
    angle = 2*pi*i/dt
    l,r = drive(angle, 5)
    print('{:6} {:6.2f} {:6.2f}'.format(angle*180/pi, l, r))
```

```python
a = []
l = []
r = []
m = []
dt = 100
speed = 1
for i in range(dt):
    angle = 2*pi*i/dt
    left, right = drive(angle, speed)
    a.append(angle)
    l.append(left)
    r.append(right)
#     m.append(sqrt(left**2+right**2))
    m.append((left+right))
    
ax = plt.subplot(121, projection='polar')
ax.scatter(a,l,label='left')
ax.scatter(a,r,label='right')
ax.scatter(a,m,label='mag')
ax.legend()
ax.grid(True);

ax = plt.subplot(122)
aa = [x*180/pi for x in a]
ax.scatter(aa,l,label='left')
ax.scatter(aa,r,label='right')
ax.legend()
ax.grid(True);
```
