```python
%matplotlib inline
```

```python
from __future__ import print_function
from __future__ import division
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# make it look nice
plt.grid(True)
plt.legend()
plt.title('title')
plt.ylabel('y')
plt.xlabel('x');  # note, if you don't put the ';' at the end, it prints 
                  # the object address. Not an issue, but it annoys me
```

```python
from math import cos

# Prepare the data
x = np.linspace(0, 10, 100)
y = [x*cos(x) for x in np.linspace(0, 10, 100)]

# similar to matlab
plt.plot(x, x, 'r-', x, y, 'b^')

# make it look nice
plt.grid(True)
# plt.legend()
plt.title('title')
plt.ylabel('y')
plt.xlabel('x');
```

```python
plt.subplot(1,2,1)
plt.plot(x, x, label='linear')
plt.title('no grid')

plt.subplot(1,2,2)
plt.plot(x, x, label='linear')
plt.title('grid');
plt.grid(True)
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75);
plt.grid(True)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # put text at (x,y)=(60,0.025)
plt.axis([40, 160, 0, 0.03]);
```
